```python
import jams
import glob
```

```python
all_chords = set()

for jam_path in glob.glob("v1.0.0/v1.0.0/jams/billboard_*.jams"):
    jam = jams.load(jam_path, validate=False)
    ann = jam["annotations"][0]["data"]

    for obs in ann:
        all_chords.add(obs.value)

# Print all unique chord labels found
for chord in sorted(all_chords):
    print(chord)
```
